```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
def sigmod(x):
    return 1/(1+np.exp(-x))

def dev_sigmod(x):
    return sigmod(x) * (1-sigmod(x))
```

```python
start = 10
x = np.linspace(-start, start, 100)
y = sigmod(x)
plt.plot(x, y)
dev = dev_sigmod(x)
plt.plot(x, dev)
plt.grid()
plt.savefig('../CNN/sigmod.jpg')
```

```python
def tanh(x):
    up = np.exp(x) - np.exp(-x)
    down = np.exp(x) + np.exp(-x)
    return up/down

def dev_tanh(x):
    return 1-tanh(x)*tanh(x)
```

```python
y = tanh(x)
dev = dev_tanh(x)
plt.plot(x, y)
plt.plot(x, dev)
plt.grid()
plt.savefig('../CNN/tanh.jpg')
```

```python
def relu(x):
    return np.max(0, x)
```

```python
y = relu(x)
plt.plot(x, y)
x1 = x[x<=0]
plt.plot(x1, 0)
x2 = x[x>0]
plt.plot(x2, 1)
```
